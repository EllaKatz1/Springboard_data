```python
#Data loading:
Phenotypes = pd.read_csv(r"C:\Users\Ella Katz\Desktop\Post-doc\Personal\Springboard\git_hub_projects\Springboard_data\Second_capstone\Datasets\All_Data_emmeans_info_V1.csv")
Classifications = pd.read_csv(r"C:\Users\Ella Katz\Desktop\Post-doc\Personal\Springboard\git_hub_projects\Springboard_data\Second_capstone\Datasets\Seeds_classification.csv")
Environmental_data = pd.read_csv(r"C:\Users\Ella Katz\Desktop\Post-doc\Personal\Springboard\git_hub_projects\Springboard_data\Second_capstone\Datasets\Environmental_data.csv")
```

```python
print(Phenotypes.shape)
print(Classifications.shape)
print(Environmental_data.shape)
```

```python
#Data joining:
Data_1 = pd.merge(Phenotypes, Classifications, how="left", on="CS_number")

print(Phenotypes.shape)
print(Classifications.shape)
print(Data_1.shape)
```

```python
Data_1.dtypes
```

```python
Data_1['Seeds_classification'].nunique()
```

```python
#Creating dummies from Seeds_classification:
Data_2 = pd.get_dummies(Data_1, prefix='Chemotype_', columns=['Seeds_classification'])
Data_3 = pd.get_dummies(Data_2, prefix='Treatment_', columns=['N_treatment'])
Data_3.head()
```

```python
#Scale standardization
from sklearn import preprocessing

min_max_scaler = preprocessing.MinMaxScaler()
Scaled = min_max_scaler.fit_transform(Data_3[['Leaf_Area', 'PR_length', 'Number_of_LR', 'Total_LR', 'Total_GSL']])
Data_4 = pd.DataFrame(Scaled)
print(Data_4.describe())
print(Data_4.shape)
```

```python
#Orgnizing and combining with the rest of the data:
Data_4.columns = ['Leaf_Area', 'PR_length', 'Number_of_LR', 'Total_LR',  'Total_GSL']
Data_4.head()
```

```python
Data_3.drop(['Leaf_Area', 'PR_length', 'Number_of_LR', 'Total_LR',  'Total_GSL'], axis=1, inplace=True)
Data_3.head()
```

```python
Data_5 = pd.concat([Data_3, Data_4], axis=1)
Data_5.shape
```

```python
#Spliting the data:
from sklearn.model_selection import train_test_split

train, test = train_test_split(Data_5, test_size=0.2)
print(train.shape)
print(test.shape)
```
